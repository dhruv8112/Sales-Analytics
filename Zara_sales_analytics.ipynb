```python
import pandas as pd
import re
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)  # Remove punctuation & numbers
    return ' '.join([word for word in text.split() if word not in stop_words])

df['clean_description'] = df['description'].apply(clean_text)
df['clean_description']
```

```python
df['tokens'] = df['clean_description'].str.split()
df_exploded = df.explode('tokens')  # Each word becomes its own row
df_exploded
```

```python
word_sales = df_exploded.groupby(['tokens','section'])['Total_sales'].agg(['count', 'sum', 'mean']).sort_values(by='sum', ascending=False)
words_tokenization=word_sales.round(1)
print(words_tokenization)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

tfidf = TfidfVectorizer(max_features=100)
X = tfidf.fit_transform(df['clean_description'])

kmeans = KMeans(n_clusters=5, random_state=0)
df['cluster'] = kmeans.fit_predict(X)
```

## Analyze sales by position/promotion within clusters

### Subtask:
Calculate and visualize the average and total sales for each 'Product Position' and 'Promotion' category within each cluster.


**Reasoning**:
Calculate the average and total sales for each 'Product Position' and 'Promotion' category within each cluster, and then visualize these results with bar plots.



```python
# Calculate sales by product position within each cluster
product_position_sales = df.groupby(['cluster', 'Product Position'])['Total_sales'].agg(['mean', 'sum']).reset_index()

# Calculate sales by promotion within each cluster
promotion_sales = df.groupby(['cluster', 'Promotion'])['Total_sales'].agg(['mean', 'sum']).reset_index()

# Visualize average sales by product position within each cluster
plt.figure(figsize=(8, 6))
sns.barplot(x='cluster', y='mean', hue='Product Position', data=product_position_sales)
plt.title('Average Sales by Product Position within Clusters')
plt.xlabel('Cluster')
plt.ylabel('Average Sales')
plt.show()

# Visualize total sales by product position within each cluster
plt.figure(figsize=(8, 6))
sns.barplot(x='cluster', y='sum', hue='Product Position', data=product_position_sales)
plt.title('Total Sales by Product Position within Clusters')
plt.xlabel('Cluster')
plt.ylabel('Total Sales')
plt.show()

# Visualize average sales by promotion within each cluster
plt.figure(figsize=(8, 6))
sns.barplot(x='cluster', y='mean', hue='Promotion', data=promotion_sales)
plt.title('Average Sales by Promotion within Clusters')
plt.xlabel('Cluster')
plt.ylabel('Average Sales')
plt.show()

# Visualize total sales by promotion within each cluster
plt.figure(figsize=(8, 6))
sns.barplot(x='cluster', y='sum', hue='Promotion', data=promotion_sales)
plt.title('Total Sales by Promotion within Clusters')
plt.xlabel('Cluster')
plt.ylabel('Total Sales')
plt.show()
```

## Explore frequent terms in descriptions

### Subtask:
Identify and display the most common words or n-grams within the 'clean_description' for each cluster to understand the textual characteristics of products in each group.


**Reasoning**:
Identify and display the most common words or n-grams within the 'clean_description' for each cluster to understand the textual characteristics of products in each group.



```python
from collections import Counter

# Loop through each cluster
for cluster_id in sorted(df['cluster'].unique()):
    words = df_exploded[df_exploded['cluster'] == cluster_id]['tokens']
    top_words = Counter(words).most_common(10)
    print(f"\nTop words in Cluster {cluster_id}:")
    for word, count in top_words:
        print(f"{word} ({count})")
```

**Reasoning**:
The error indicates that the 'cluster' column does not exist in the `df_exploded` DataFrame. I need to merge the cluster information from the original `df` DataFrame into `df_exploded`.



```python
# Merge the cluster information into the exploded DataFrame
df_exploded = df_exploded.merge(df[['Product ID', 'cluster']], on='Product ID', how='left')

# Loop through each cluster
from collections import Counter

for cluster_id in sorted(df_exploded['cluster'].unique()):
    words = df_exploded[df_exploded['cluster'] == cluster_id]['tokens']
    top_words = Counter(words).most_common(10)
    print(f"\nTop words in Cluster {cluster_id}:")
    for word, count in top_words:
        print(f"{word} ({count})")
```

## Investigate seasonal products

### Subtask:
Analyze and visualize the distribution of 'Seasonal' products across clusters and potentially their sales performance within each cluster.


**Reasoning**:
Calculate the count of seasonal and non-seasonal products within each cluster and visualize the distribution. Then calculate and visualize the average total sales for seasonal and non-seasonal products within each cluster.



```python
# Calculate the count of 'Seasonal' and 'Non-Seasonal' products within each cluster
seasonal_counts = df.groupby(['cluster', 'Seasonal']).size().reset_index(name='count')
print("Seasonal counts per cluster:\n", seasonal_counts)

# Visualize the distribution of 'Seasonal' products across the different clusters
plt.figure(figsize=(8, 6))
sns.barplot(x='cluster', y='count', hue='Seasonal', data=seasonal_counts, palette='viridis')
plt.title('Distribution of Seasonal Products Across Clusters')
plt.xlabel('Cluster')
plt.ylabel('Number of Products')
plt.show()

# Calculate the average 'Total_sales' for 'Seasonal' and 'Non-Seasonal' products within each cluster
seasonal_sales_mean = df.groupby(['cluster', 'Seasonal'])['Total_sales'].mean().reset_index(name='Average Total Sales')
print("\nAverage total sales for seasonal products per cluster:\n", seasonal_sales_mean)

# Create a bar plot to visualize the average 'Total_sales' of seasonal and non-seasonal products per cluster
plt.figure(figsize=(8, 6))
sns.barplot(x='cluster', y='Average Total Sales', hue='Seasonal', data=seasonal_sales_mean, palette='viridis')
plt.title('Average Total Sales of Seasonal vs Non-Seasonal Products per Cluster')
plt.xlabel('Cluster')
plt.ylabel('Average Total Sales')
plt.show()
```
